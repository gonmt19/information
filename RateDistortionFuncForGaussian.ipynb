```python
import numpy as np
from math import log2
from matplotlib import pyplot as plt


def rdf(sigma, d):
    """ rate distortion function for Gaussigna source

    Args:
        sigma: variance
        d: expected distorion constraint
    """
    if d > sigma*sigma:
        return 0
    else:
        return log2(sigma*sigma/d) / 2


if __name__ == "__main__":
    sigma_arr = [0.5, 1.0, 5.0]
    d_arr = np.arange(0.1, 30., 0.05)
    fig, ax = plt.subplots(figsize=(5, 5))
    for sigma in sigma_arr:
        rdf_val = []
        for d in d_arr:
            rdf_val.append(rdf(sigma, d))
        ax.plot(d_arr, rdf_val, label=r"$\sigma$: %.1f" % sigma)
    ax.set_title(r"Rate distortion function for N(0, $\sigma^2$)")
    ax.set_xlabel("d")
    ax.set_ylabel("R(d)")
    ax.grid(True)
    ax.legend()
    plt.show()
```
