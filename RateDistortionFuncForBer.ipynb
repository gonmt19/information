```python
import numpy as np
from matplotlib import pyplot as plt
from math import log2


def entropy(p):
    if p == 0 or p == 1:
        return 0
    return -p * log2(p) - (1 - p) * log2(1 - p)


def rdf(p, d):
    """rate distortion function

    Args:
        p: Bernulli parameter
        d: expected distorion constraint
    """
    if d <= min(p, 1 - p):
        return entropy(p) - entropy(d)
    return 0


if __name__ == "__main__":
    p_arr = [0.2, 0.5, 0.8, 1.0]
    d_arr = np.arange(0.0, 1.0, 0.01)
    fig, ax = plt.subplots(figsize=(10, 10))
    for p in p_arr:
        rdf_val = []
        for d in d_arr:
            rdf_val.append(rdf(p, d))
        ax.plot(d_arr, rdf_val, label="p : %.2f" % p)
    ax.set_title("rate distortion function for Ber(p)")
    ax.grid(True)
    ax.legend()
    plt.show()
```

rate distortion function is same when p and 1 - p
